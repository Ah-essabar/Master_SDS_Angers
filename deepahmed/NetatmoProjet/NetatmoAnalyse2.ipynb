```python
#!pip install statsmodels
#!pip install seaborn
from pandas import read_csv
from matplotlib import pyplot
import seaborn as sns
from statsmodels.tsa.seasonal import seasonal_decompose
series = read_csv('DonnesNetatmo1mai.csv',sep=';', header=0, index_col=0)
```

```python
s=series[["co2"]]
pyplot.plot(s)
pyplot.show()
```

```python
result = seasonal_decompose(s, model='multiplicative',period=300)
result.plot()
pyplot.show()
```
