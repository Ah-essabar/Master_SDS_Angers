```python
def lagged(df,lag=2):
    #df=df.set_index('time')    
    Df=[]
    a=[]
    for i in range(lag):
        a.append(i)
        df1=df.shift(i+1).copy()
        df1=df1.add_suffix('_t-'+str(i+1))
        Df.append(df1)
    for base in Df:
        df=pd.concat([df,base], axis=1)
    
    #df=df.drop(df.index[a])
    return df, a 
```
