```python
listefile = ["05.csv","10.csv","25.csv","40.csv", "60.csv", "70.csv"]
for p in listefile:
    df=pd.read_csv('BDD_missingData/MCAR'+p, sep=",", encoding = "ISO-8859-1", index_col=index_col)
    for col in df.columns:
        df[col]
        mask = df[col].isna()
        to_fill = np.random.uniform(df[col].min(), df[col].max(), size=mask.sum())
        df.loc[mask, col] = to_fill
  
    df.to_csv("./ImputedData/RandomImputer_MCAR"+p)

for p in listefile:
    df=pd.read_csv('BDD_missingData/MAR'+p, sep=",", encoding = "ISO-8859-1", index_col=index_col)
    for col in df.columns:
        df[col]
        mask = df[col].isna()
        to_fill = np.random.uniform(df[col].min(), df[col].max(), size=mask.sum())
        df.loc[mask, col] = to_fill
  
    df.to_csv("./ImputedData/RandomImputer_MAR"+p)    
```
